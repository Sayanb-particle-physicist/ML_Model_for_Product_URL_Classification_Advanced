# **Train/test split**

```python
from sklearn.model_selection import train_test_split

X = df['url'].values
y = df['label_bin'].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=42
)
```

# **URL tokenizer**

```python
import re
def url_tokenizer(u):
    u = u.lower()
    u = re.sub(r'^https?://', '', u)
    tokens = re.split('[^a-z0-9]+', u)
    tokens = [t for t in tokens if t and not t.isdigit()]
    return tokens
```

# **Meta features extractor**

```python
from urllib.parse import urlparse
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin
from scipy.sparse import csr_matrix

class MetaFeatures(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        X = np.array(X).ravel()
        rows = []
        for u in X:
            parsed = urlparse(u)
            path = parsed.path or ""
            query = parsed.query or ""
            netloc = parsed.netloc or ""
            rows.append([
                len(u),
                path.count('/'),
                sum(c.isdigit() for c in u),
                1 if query else 0,
                1 if '/dp/' in u or '/gp/product/' in u else 0,
                1 if re.search(r'/p/|/itm', u) else 0,
                1 if 'amazon.' in netloc else 0,
                1 if 'flipkart.' in netloc else 0,
                1 if 'google.' in netloc else 0,
                1 if 'myntra.' in netloc else 0
            ])
        return csr_matrix(np.array(rows, dtype=float))
```

# **Build and train pipeline**

```python
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MaxAbsScaler
from sklearn.linear_model import LogisticRegression

tfidf_pipe = Pipeline([
    ('vect', TfidfVectorizer(
        tokenizer=url_tokenizer,
        token_pattern=None,
        ngram_range=(1,2),
        max_features=1500
    ))
])


meta_pipe = Pipeline([
    ('meta_feats', MetaFeatures()),
    ('scaler', MaxAbsScaler())
])

union = FeatureUnion([
    ('tfidf', tfidf_pipe),
    ('meta', meta_pipe)
])

pipeline = Pipeline([
    ('features', union),
    ('clf', LogisticRegression(max_iter=2000, class_weight='balanced', random_state=42))
])

pipeline.fit(X_train, y_train)
```

# **Evaluation of the Model**

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

y_pred = pipeline.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=['not_product','product']))
print(confusion_matrix(y_test, y_pred))
```
